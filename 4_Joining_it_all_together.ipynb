```python
import pandas as pd
import numpy as np

binary = pd.read_csv('Data/binary_train_selected.csv')
ordinal = pd.read_csv('Data/ordinal_train_selected.csv')
numerical = pd.read_csv('Data/num_train_selected.csv')
time = pd.read_csv('Data/time_train_selected.csv')

full_data = pd.concat([binary, ordinal, numerical, time], axis=1).drop('Unnamed: 0', axis=1)


full_data
```

# Test and Train data split

```python
np.random.seed(42)

N = len(full_data)
idx = np.arange(0, N)
np.random.shuffle(idx)

train = full_data.iloc[idx[:int(0.8*N)]]
test = full_data.iloc[idx[int(0.8*N):]]

train
```

## Dummy Regressor
As a reference point for training and test data

```python
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_squared_error

dummy_r = DummyRegressor(strategy='mean')


dummy_r.fit(train.drop(['SalePrice'], axis=1), train['SalePrice'])

pred = dummy_r.predict(test.drop(['SalePrice'], axis=1))

MSE =  mean_squared_error(pred, test['SalePrice'])

print(f'Dummy regressor (calculated mean) got a score: {MSE=}')
```

## Save

```python
train.to_csv('Data/final_train.csv')
test.to_csv('Data/final_test.csv')
print("Data Saved")
```
